```python
df=df.drop('id',axis=1)
```

```python
df['data'] =pd.to_datetime(df['date'])
```

```python
df['data']
```

```python
#here This date time by itself may not be useful, ****but we can extract the year component and the month component.****

#And the way I can do that is through a couple of simple calls.
```

```python
#def year_extraction(date):        #==  lambda date: date.year
    #return  date.year
```

```python
df['year']=df['data'].apply(lambda date: date.year)
df['month']=df['data'].apply(lambda date: date.month)
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='month', y='price',data=df)
```

```python
df.groupby('month')['price'].mean()
```

```python
df.groupby('month')['price'].mean().plot()
```

```python
df.groupby('year')['price'].mean().plot()


#hey're just increasing in price as time goes on.
#And you would expect that just naturally with inflation, unless there's some sort of major housing events.
```
